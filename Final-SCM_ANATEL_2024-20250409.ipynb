```
%sql DROP DATABASE if exists prata CASCADE;
```

```
%sql CREATE DATABASE prata;
```

```
%sql create or replace view prata.vw_acessos_scm as
select Ano_Mes_Oper_codIBGE, UF, Municipio , sum(Acessos) as num_acessos
from bronze.acessos_scm
where Ano = '2024'
and Empresa in ('TIM','CLARO','VIVO')
and Tipo_de_Produto = 'INTERNET'
and Cod_IBGE in ('1302603','5300108','3550308','3304557','5208707','2927408','3106200') -- 7 capitais selecionadas (Manaus, Brasilia, São Paulo, Rio de Janeiro, Goiânia, Salvador e Belo Horizonte)
group by Ano_Mes_Oper_codIBGE, UF, Municipio
```

```
%sql
select * from prata.vw_acessos_scm
limit 4
```

```
%sql 
create or replace view prata.vw_indicadores_rede_scm as
select Ano_Mes_Oper_codIBGE, Ano, Mes, Indicador, Descricao, Cod_IBGE, Municipio, UF, Prestadora, Resultado,Estatistica, Nro_de_Coletores, Nro_de_Medidas
from bronze.indicadores_rede_scm
where Servico = 'Banda Larga Fixa'
and Prestadora in ('TIM','CLARO','VIVO')
and Ano = '2024'
and tipo = "Indicador IQS"
and indicador like "IND%"
and Cod_IBGE in ('1302603','5300108','3550308','3304557','5208707','2927408','3106200') -- 7 capitais selecionadas (Manaus, Brasilia, São Paulo, Rio de Janeiro, Goiânia, Salvador e Belo Horizonte)
```

```
%sql
select *
from prata.vw_indicadores_rede_scm
where Municipio = 'Salvador'
and Mes = '12'
and  Prestadora = 'VIVO'
limit 15
```

```
%sql 
create or replace view prata.vw_reclamacoes_scm as
select Ano_Mes_Oper_codIBGE, SUM(solicitacoes) as num_solicitacoes, Ano, Mes, UF, Cidade, Marca 
from bronze.reclamacoes_scm
where Servico = 'SCM'
and Marca in ('TIM','CLARO','VIVO')
and Ano = '2024'
and Cod_IBGE in ('1302603','5300108','3550308','3304557','5208707','2927408','3106200') -- 7 capitais selecionadas (Manaus, Brasilia, São Paulo, Rio de Janeiro, Goiânia, Salvador e Belo Horizonte)
and Assunto in ('QUALIDADE, FUNCIONAMENTO E REPARO','INSTALAÇÃO OU ATIVAÇÃO OU HABILITAÇÃO','MUDANÇA DE ENDEREÇO')
group by Ano_Mes_Oper_codIBGE, Ano, Mes, UF, Cidade, Marca
```

```
%sql
select * from prata.vw_reclamacoes_scm
where marca = 'TIM'
and mes = '12'
limit 10
```

```
%sql DROP DATABASE if exists ouro CASCADE;
```

```
%sql CREATE DATABASE ouro
```

```
%sql 
create or replace view ouro.anatel_scm as 

SELECT I.*, A.num_acessos, R.num_solicitacoes
FROM prata.vw_indicadores_rede_scm  I
     LEFT JOIN prata.vw_acessos_scm A on A.Ano_Mes_Oper_codIBGE = I.Ano_Mes_Oper_codIBGE
     LEFT JOIN prata.vw_reclamacoes_scm R on I.Ano_Mes_Oper_codIBGE = R.Ano_Mes_Oper_codIBGE
     
```

```
%sql
select * from ouro.anatel_scm
where Municipio = 'Salvador'
and Mes = '3'
and  Prestadora = 'VIVO'
limit 15
```
